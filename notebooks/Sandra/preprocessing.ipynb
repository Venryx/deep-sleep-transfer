```python
import os
```

```python
# delete unnecessary large files
folder = "D:/physionet_challenge/processed/training/iitnet_0/"
subfolders = [ f.path for f in os.scandir(folder) if f.is_dir() ]
print("found", str(len(subfolders)), "folders in directory.")
```

```python
for folder in subfolders:
    x_file = folder + "/0_data_x_preprocessed.npy"
    print(x_file)
    if os.path.exists(x_file):
        os.remove(x_file)
        print("deleted")
    else:
        print("does not exist!")
    
    y_file = folder + "/0_data_y_preprocessed.npy"
    print(y_file)
    if os.path.exists(y_file):
        os.remove(y_file)
        print("deleted")
    else:
        print("does not exist!")
```

```python
# put 199 randomly selected samples from training to test folder
import random
test_subjects = random.sample(subfolders, 199)
print("Selected", str(len(test_subjects)), "subjects.")
```

```python
import shutil
dest_folder = "D:/physionet_challenge/processed/test/iitnet_0/"

for test_subject in test_subjects:
    print("Moving", test_subject, "to", dest_folder)
    shutil.move(test_subject, dest_folder)
    print("done.")

print("Moved", str(len(test_subjects)), "subjects.")
```
